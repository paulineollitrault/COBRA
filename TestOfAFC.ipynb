```python
import numpy as np
import time
from qiskit.aqua.operators import WeightedPauliOperator, commutator
from qiskit.chemistry import FermionicOperator
import AnalyticFermionicCommutator as AFC
```

```python
nferm = 4

told = []
tnew = []

for nferm in range(4,20):
    print('# ferm->', nferm)
    h1 = np.zeros((nferm,nferm))
    h1[1,2] = 1
    h1[2,1] = 1
    h2 = np.zeros((nferm,nferm,nferm,nferm))
    h2[1,2,1,3] = 1
    h2[2,1,1,3] = -1
    h11 = np.zeros((nferm,nferm))

    h2c = np.einsum('ikmj->ijkm', h2)

    t0 = time.time()
    ferop = FermionicOperator(h1)
    ferop2 = FermionicOperator(h11,h2c)
    A = ferop.mapping('jordan_wigner')
    B = ferop2.mapping('jordan_wigner')
    result = commutator(A,B)
    t1 = time.time()
    print('Computational time of QisKit routine:',t1-t0)
    told.append(t1-t0)
    
    t0 = time.time()
    A = AFC.h_matrix_to_terms(h1)
    B = AFC.h_matrix_to_terms(h2)
    Com = AFC.FermCommutator(A,B)
    H = AFC.oper_to_h(Com,nferm)
    hafc = H[0]
    hc = np.einsum('ikmj->ijkm', hafc)
    ferop3 = FermionicOperator(h11,hc)
    Q = ferop3.mapping('jordan_wigner')
    t1 = time.time()
    print('Computational time of new routine:',t1-t0)
    tnew.append(t1-t0)
    
    if 2*Q == result:
        print('OK')    
```

```python
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 18, 10
rcParams.update({'font.size': 26})
plt.xlabel('number of qubits',fontsize=28)
plt.ylabel('Computational time',fontsize=28, color='black')

plt.plot(range(4,20),told, label='Qiskit')
plt.plot(range(4,20),tnew, label='New')
plt.legend()
plt.show()
plt.savefig('results.png')
```
